# Python Special Methods


Dunder = **D**ouble **UNDER**score. Like `__init__` or `__str__`.

These methods let your classes work with Python's built-in stuff like `print()`, `len()`, `==`, and loops.

## 1. `__init__` — Set Up Your Object

Runs automatically when you create an object. Use it to set starting values.

```python
class Bird:
    def __init__(self, name, wingspan):
        self.name = name
        self.wingspan = wingspan

eagle = Bird("Eagle", 2.3)
print(f"{eagle.name} has a {eagle.wingspan}m wingspan")
```

## 2. `__str__` and `__repr__` — String Output

- `__str__`: What you see when you `print()` the object
- `__repr__`: What developers see (should show how to recreate the object)

```python
class Album:
    def __init__(self, title, artist):
        self.title = title
        self.artist = artist
    
    def __str__(self):
        return f"{self.title} by {self.artist}"
    
    def __repr__(self):
        return f"Album('{self.title}', '{self.artist}')"

a = Album("Thriller", "Michael Jackson")
print(str(a))   # For users
print(repr(a))  # For devs
```

## 3. `__len__` — Make `len()` Work

Lets you use `len()` on your object. Must return an integer.

```python
class Toolbox:
    def __init__(self):
        self.tools = []
    
    def add(self, tool):
        self.tools.append(tool)
    
    def __len__(self):
        return len(self.tools)

box = Toolbox()
box.add("hammer")
box.add("screwdriver")
box.add("wrench")
print(f"Tools: {len(box)}")
```

## 4. `__eq__` — Compare with `==`

By default, `==` checks if two variables point to the same object.
With `__eq__`, you can compare by content instead.

```python
class Member:
    def __init__(self, id, name):
        self.id = id
        self.name = name
    
    def __eq__(self, other):
        return self.id == other.id

m1 = Member(1, "Alice")
m2 = Member(1, "Alice")
print(f"Same content: {m1 == m2}")   # True
print(f"Same object: {m1 is m2}")    # False
```

## 5. `__add__`, `__sub__`, `__mul__` — Math Operators

Teach Python what `+`, `-`, `*` mean for your objects.

```python
class Matrix:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    
    def __add__(self, other):
        return Matrix(
            self.a + other.a, self.b + other.b,
            self.c + other.c, self.d + other.d
        )
    
    def __sub__(self, other):
        return Matrix(
            self.a - other.a, self.b - other.b,
            self.c - other.c, self.d - other.d
        )
    
    def __mul__(self, scalar):
        return Matrix(
            self.a * scalar, self.b * scalar,
            self.c * scalar, self.d * scalar
        )
    
    def __str__(self):
        return f"[{self.a} {self.b}; {self.c} {self.d}]"

m1 = Matrix(1, 2, 3, 4)
m2 = Matrix(5, 6, 7, 8)
print(f"m1 + m2 = {m1 + m2}")
print(f"m1 - m2 = {m1 - m2}")
print(f"m1 * 2 = {m1 * 2}")
```

## 6. `__getitem__` — Index with `[]`

Lets you use bracket notation on your objects: `obj[0]`, `obj[1]`, etc.

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def enqueue(self, item):
        self.items.append(item)
    
    def __getitem__(self, index):
        return self.items[index]

q = Queue()
q.enqueue("first")
q.enqueue("second")
q.enqueue("third")
print(q[0])
print(q[-1])
```

## 7. `__setitem__` — Assign with `[]`

Lets you do `obj[0] = value`.

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def enqueue(self, item):
        self.items.append(item)
    
    def __getitem__(self, index):
        return self.items[index]
    
    def __setitem__(self, index, value):
        self.items[index] = value

q = Queue()
q.enqueue("old")
q.enqueue("another")
print(f"Before: {q.items}")
q[0] = "new"
print(f"After:  {q.items}")
```

## 8. `__contains__` — The `in` Operator

Defines what happens when you use `in` on your object.

```python
class Cupboard:
    def __init__(self):
        self.items = []
    
    def put(self, item):
        self.items.append(item)
    
    def __contains__(self, item):
        return item in self.items

c = Cupboard()
c.put("plates")
c.put("cups")
print(f"Got plates? {'plates' in c}")
print(f"Got bowls? {'bowls' in c}")
```

## 9. `__call__` — Call Object Like a Function

Use `obj()` to call your object. Good for things that need settings but act like functions.

```python
class Converter:
    def __init__(self, rate):
        self.rate = rate
    
    def __call__(self, amount):
        return amount * self.rate

usd_to_eur = Converter(0.85)
usd_to_gbp = Converter(0.73)
print(f"$100 = €{usd_to_eur(100):.2f}")
print(f"$100 = £{usd_to_gbp(100):.2f}")
```

## 10. `__iter__` and `__next__` — Loop with `for`

Makes your object work in `for` loops.

- `__iter__` returns the iterator (usually `self`)
- `__next__` returns the next value or raises `StopIteration` when done

```python
class Stepper:
    def __init__(self, start, end, step):
        self.start = start
        self.end = end
        self.step = step
    
    def __iter__(self):
        self.current = self.start
        return self
    
    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        num = self.current
        self.current += self.step
        return num

for n in Stepper(0, 10, 2):
    print(n, end=" ")
print("Done!")
```

## 11. `__bool__` — True or False?

Controls whether your object is truthy or falsy in `if` statements.

```python
class Vault:
    def __init__(self, gold):
        self.gold = gold
    
    def __bool__(self):
        return self.gold > 0

v1 = Vault(100)
v2 = Vault(0)
print(f"Has gold: {bool(v1)}")
print(f"Has gold: {bool(v2)}")
```

## 12. `__del__` — Cleanup (Optional)

Called when an object is about to be destroyed.

**Warning:** Don't rely on this for important cleanup. Use `with` statements instead.

```python
class TempLog:
    def __init__(self, name):
        self.name = name
        print(f"Opened: {name}")
    
    def __del__(self):
        print(f"Closed: {self.name}")

log = TempLog("debug.log")
print("Writing...")
del log
print("Done.")
```

| `__repr__` | Debug info | `repr(obj)` | `__repr__` |
| `__len__` | Length | `len(obj)` | `__len__` |
| `__eq__` | Equal check | `obj1 == obj2` | `__eq__` |
| `__add__` | Addition | `obj1 + obj2` | `__add__` |
| `__sub__` | Subtraction | `obj1 - obj2` | `__sub__` |
| `__mul__` | Multiply | `obj * 3` | `__mul__` |
| `__getitem__` | Get item | `obj[0]` | `__getitem__` |
| `__setitem__` | Set item | `obj[0] = x` | `__setitem__` |
| `__contains__` | Check `in` | `x in obj` | `__contains__` |
| `__call__` | Call object | `obj()` | `__call__` |
| `__iter__` | Make iterable | `for x in obj` | `__iter__` |
| `__next__` | Next item | `next(obj)` | `__next__` |
| `__bool__` | True/False | `if obj:` | `__bool__` |
| `__del__` | Cleanup | `del obj` | `__del__` |

# Remember

1. You don't call these methods directly — Python does it for you.
2. `__init__` sets up your object when it's created.
3. `__str__` is for showing to users; `__repr__` is for debugging.
4. `__eq__` lets you compare objects by their content, not just memory address.
5. Operator methods (`__add__`, etc.) make your types feel natural to use.
6. `__getitem__`/`__setitem__` let you use `[]` on your objects.
7. `__iter__`/`__next__` make your objects work in `for` loops.
8. `__call__` makes objects act like functions.
9. `__bool__` controls whether your object is truthy or falsy.
10. Don't rely on `__del__` for cleanup — use `with` statements instead.

**The main idea:** If Python has a built-in way to interact with data (print, add, compare, loop), there's a dunder method that lets your class join the party.
